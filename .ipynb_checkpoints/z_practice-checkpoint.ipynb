```python
import random
import numpy as np

class KMeans:
    def __init__(self,n_clusters=4,max_iter=100):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.centroids = None

    def fit_predict(self,X):
        X = X.values

        random_index = random.sample(range(0,X.shape[0]),self.n_clusters)
        # randomly picking k points from data set , 
        # k is number of clusters
                
        self.centroids = X[random_index]
        # assigning random k points as centroids 
        

        for i in range(self.max_iter):
            # assign clusters
            cluster_group = self.assign_clusters(X)
            # each iteration , assigning clusters
            
            old_centroids = self.centroids
            # moving centroids
            self.centroids = self.move_centroids(X,cluster_group)
            # check finish
            if (old_centroids == self.centroids).all():
                break

        return cluster_group

    def assign_clusters(self,X):
        cluster_group = []
        distances = []

        for r in X: # for each data point
            for c in self.centroids:   # for each centroids
                distances.append(np.linalg.norm(r-c))  # find distnace between centroid and data points
            min_distance = min(distances)   # get minimum distance
            index_pos = distances.index(min_distance)  # which centroid is closet distance 
            cluster_group.append(index_pos) # assign the cluster 
            distances.clear()
        # repeating the entire process for each data points and each clusters.
        return np.array(cluster_group)

    def move_centroids(self,X,cluster_group):
        new_centroids = []

        cluster_type = np.unique(cluster_group)

        for type in cluster_type:
            new_centroids.append(X[cluster_group == type].mean(axis=0))
            # moving centroids , taking mean of eacch cluster type 
            # for each iteration  
        return np.array(new_centroids)
```

```python

import matplotlib.pyplot as plt
def plot(data, centroids):
    plt.scatter(data[:, 0], data[:, 1], marker = '.',
                color = 'gray', label = 'data points')
    plt.scatter(centroids[:, 0], centroids[:, 1],
                color = 'red', label = 'centroid')     
    plt.legend()
    plt.xlim(-5, 12)
    plt.ylim(-10, 15)
    plt.show()
```

```python
def geting_all_distances(X,centroids):
    dist = np.array([]).reshape(len(X),0)
    for c in centroids:
        d = np.linalg.norm(X - c, axis = 1).reshape(-1,1)
        dist = np.hstack([dist,d])
    return dist
        
```

```python
geting_all_distances(X,centroids)
```

```python
def get_the_closet_centroid(X,centroids):
    return np.argmin(geting_all_distances(X,centroids),axis = 1)
```
