```python
df1 = pd.read_csv("Stocks.csv",index_col=0)
```

```python
df1
```

```python
df1.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
df = std.fit_transform(df1)
```

```python
df
```

```python
import scipy.cluster.hierarchy as sch
```

```python
z = sch.linkage(df, method='ward', metric='euclidean')
```

```python
z.
```

```python
plt.figure(figsize=(15,10))
sch.dendrogram(z, labels = df1.index,color_threshold=2)
plt.xticks(rotation=90)
plt.ylabel('distance')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sch.dendrogram(z, labels = df1.index,color_threshold=4)
plt.xticks(rotation=90)
plt.ylabel('distance')
plt.show()
```

# customer_segmentation

```python
df = pd.read_csv('customer_segmentation.csv')
X = df.copy()
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering


hc = AgglomerativeClustering(
    n_clusters=4,
    affinity='euclidean',
    memory=None,
    connectivity=None,
    compute_full_tree='auto',
    linkage='ward',
    distance_threshold=None,
    compute_distances=False,)
```

```python
hc.fit_predict(X)
```

```python
df["labels"] = hc.fit_predict(X)
```

```python
df
```

```python
df.groupby("labels")["n_clicks"].mean().plot(kind = "bar")
```

```python
df.groupby("labels")["n_visits"].mean().plot(kind = "bar")
```

```python
df.groupby("labels")["amount_spent"].mean().plot(kind = "bar")
```

```python
df.groupby("labels")["amount_discount"].mean().plot(kind = "bar")
```

```python
df.groupby("labels")["days_since_registration"].mean().plot(kind = "bar")
```
